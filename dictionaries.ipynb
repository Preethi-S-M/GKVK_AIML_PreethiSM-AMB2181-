```python
# shallow copy
student={"name":"Preethi","roll_no":"AMB2181","branch":"bsc agri","marks":92}
print("Original dictionary",student)

student_copy=student.copy()
print("copied dictionary",student_copy)

# to remove value
remove_value=student.pop("marks")
print("removed value",remove_value)
print("After pop",student)

# to remove last item
last_item=student.popitem()
print("Removed item",last_item)
print("after popitem",student)


# to update an item
student.update({"year":4,"branch":"agri"})
print("after update",student)
```

```python
#Nested Dictionaries

student = {
    "name": "Raj",
    "marks": {"math": 90, "cs": 95}
}
y=student["name"]
x=student["marks"]["math"]
print(y,"'s","marks in math :",x)
```

```python
# using zip function
num={1,2,3,4}
name={"sam","john","karan"}
mark={96,95,98,97}
pairs=zip(num,name,mark)
print(tuple(pairs))
for nums,names,marks in zip(num,name,mark):
  print(nums,names,marks)
```

```python
name=["sam","john","karan"]
mark=[96,95,98,97]
for sl_no,(names,marks) in enumerate(zip(name,mark),start=1):
  print(f"Student {sl_no} {names} has scored {marks}")
```
