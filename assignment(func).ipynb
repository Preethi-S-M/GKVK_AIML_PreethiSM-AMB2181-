```python
# crop yield management system
def collect_yields():
  yields=[]
  num=int(input("enter the number of fields:"))
  for i in range(num):
    value=int(input(f"enter the yield-{i+1}:"))
    yields.append(value)
  return yields

data=collect_yields()
print("Recorded crop yields:",data)

def average_yield(yields):
  total=sum(yields)
  return total/len(yields)

result=average_yield(data)
print("Average crop yield:",result)

def high_yield_crop(yields,threshold):
  high=[]
  for value in yields:
    if value>=threshold:
      high.append(value)
  return high

crop_yield=high_yield_crop(data,1000)
print("High yielding crops:",crop_yield)

def yield_statistics(yields):
  print("Minimum yield:",min(yields))
  print("Maximum yield:",max(yields))
  print("Total field:",len(yields))
  print("Average yield:",sum(yields)/len(yields))

yield_statistics(data)
```
