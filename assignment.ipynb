```python
# crop data
crops=int(input("Enter number of crops:"))
crops_name=[]
yield_per_acre=[]
for crop in range(1,crops+1):
  crop_name=input("Enter crop name:")
  crops_name.append(crop_name)
  yields=int(input("Enter yield per acre:"))
  yield_per_acre.append(yields)
print("total crops",crops_name)
print("Yield per acre:",yield_per_acre)
area=10
for crop_name,yield_per_acre in zip(crops_name,yield_per_acre):
  print(f"total yield of {crop_name}:{yield_per_acre*area}")

#rainfall data
rainfall=[]
month=int(input("Enter number of months:"))
for month_idx in range(1,month+1):
  monthly_rainfall=int(input("Enter monthly rainfall(in mm):"))
  rainfall.append(monthly_rainfall)
y=tuple(rainfall)
print("Total rainfall",rainfall)
highest=max(rainfall)
lowest=min(rainfall)
print("Highest rainfall:",highest)
print("Lowest rainfall:",lowest)
avg=sum(rainfall)/month
print("Average rainfall:",avg)
print("\n")
price_per_quintal=[]
for crop in range(1,crops+1):
  prices=int(input(f"Enter price per quintal of {crop}:"))
  price_per_quintal.append(prices)
print("Price per quintal:",price_per_quintal)
for crop,prices in zip(crops_name,price_per_quintal):
  print(f"{crop}:{prices}")
```
