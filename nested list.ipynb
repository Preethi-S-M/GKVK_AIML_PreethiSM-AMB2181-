```python
#prompting user to enter value
def expenses_made():
  expenses=[]
  number_of_days=int(input("enter number of days:"))
  for i in range(number_of_days):
    day_name=input("eneter the day name:")
    seed_cost=int(input("Enter the seed cost:"))
    fertilizer_cost=int(input("Enter the fertilizer cost:"))
    labour_cost=int(input("Enter the labour cost:"))
    expenses.append([day_name,[seed_cost,fertilizer_cost,labour_cost]])
  return expenses
total_expenditures=expenses_made()
for expenditure in total_expenditures:
  print(expenditure[0],"expenses are",expenditure[1])
```
