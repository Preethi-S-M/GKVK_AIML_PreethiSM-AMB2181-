```python
def filter_positive(lst):
  result=[]
  for num in lst:
    if num>0:
      result.append(num*2)
  return result
numbers=[1,2,-3,4]
filtered_numbers=filter_positive(numbers)
print(filtered_numbers)
```

```python
my_list=[1,2,3]
def double_pairs(lst):
  return [(x,x*2) for x in lst if x%2==1]
ourlist=double_pairs(my_list)
print(ourlist)
```

```python
def analyze(lists):
  empty_count=0
  i=0
  while i<len(lists):
    if lists[i]:
      empty_count+=1
    i+=1
  return empty_count
print(analyze([[1],[],[2,3]]))
```
