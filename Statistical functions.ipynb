```python
import numpy as np
# rainfall data(mm) of 12 months
rainfall=np.array([45,52,68,120,180,220,195,160,110,75,58,40])

# corresponding crop yield(tond/ha)
yields=np.array([2.1,2.3,2.8,4.2,5.8,6.5,5.9,5.2,3.8,2.9,2.4,2.0])

print("===RAINFALL STATISTICS=== ")
print(f"Mean:{np.mean(rainfall):.2f}mm")
print(f"median:{np.median(rainfall):.2f}mm")
print(f"Standard deviation:{np.std(rainfall):.2f}mm")
print(f"varience:{np.var(rainfall):.2f}mm")
print(f"Min:{np.min(rainfall)}mm (Month{np.argmin(rainfall)+1})")
print(f"max:{np.max(rainfall)}mm (Month{np.argmax(rainfall)+1})")
print(f"range:{np.ptp(rainfall)}mm")
print("\n")
print("===Yield staistics===")
print(f"Mean:{np.mean(yields):.2f}tons/ha")
print(f"Median:{np.median(yields):.2f} tons/ha")
print(f"Standard deviation:{np.std(yields):.2f}tons/ha")
print('\n')

print("===Percentiles===")
percentiles=[25,50,75,90]
rainfall_percentiles=np.percentile(rainfall,percentiles)
for p,val in zip(percentiles,rainfall_percentiles):
  print(f"{p}th percentile rainfall:{val:.2f}mm")

print("\n")
print('==correlation Analysis===')
correlation_matrix=np.corrcoef(rainfall,yields)
correlation=correlation_matrix[0,1]
print(f"Rainfall-yield correlation:{correlation:.4f}")
print("Interpretation: strong positive correlation" if correlation>0.7 else "moderate correlation " if correlation >0.4 else "weak correlation")

print("\n ===== cumulative rainfall===")
cumulative=np.cumsum(rainfall)
print("month\trainfall\tcumulative\t")
for month,(rain,cum) in enumerate(zip(rainfall,cumulative),start=1):
 print(f"{month}\t{rain}\t\t{cum}")
```

```python
import numpy as np
no_of_months=int(input("Enter the number of months:"))
rainfall=[]
monthyield=[]

for i in range(1,no_of_months+1):
  rains=int(input(f"Enter the rainfall of month {i}(in mm):"))
  rainfall.append(rains)
  yields=float(input(f"Enter the yield of month{i}(in q/ha):"))
  monthyield.append(yields)
total_rainfall=np.array(rainfall)
total_yield=np.array(monthyield)
print(total_rainfall)
print(total_yield)


```

```python
mode=np.bincount(total_rainfall).argmax()
print(mode)

"correlation analysis"
correaltion_matrix=np.corrcoef(total_rainfall,total_yield)
correlation=correaltion_matrix[0,1]
print(f'{correlation:.4f}')
# covarience analysis
cov=np.cov(total_rainfall,total_yield)
print(f"{cov[0,1]:.4f}")
```

```python
ptp=np.ptp(total_rainfall)
print(ptp)
argmin=np.argmin(total_rainfall)
argmax=np.argmax(total_rainfall)
print(argmin,argmax)
```

```python
# percentiles and quantiles
sorted=np.sort(total_rainfall)
print(sorted)
q25=np.percentile(total_rainfall,25)
print(q25)
percentile=[25,50,75,90]
rainfall_percentile=np.percentile(total_rainfall,percentile)
for p,val in zip(percentiles,rainfall_percentile):
  print(f"{p}th percentile rainfall:{val:.2f}mm")
```

```python
quantile=np.quantile(total_rainfall,[0.25,0.50,0.75,0.90])
print(quantile)
```

```python
"""advanced statistics"""

weighted_average=np.average(total_rainfall,weights=[1,2,3,4,2,5,4,3,6,7,2,4])
print(weighted_average)

abd=np.array([5,10,15])
weighted_avg=np.average(abd,weights=[1,2,3])
""" [(5*1)+(10*2)+(15*3)]/(1+2+3)
     (5+20+45)/6
     answer is 11.6666...~11.6"""
print(f'weighted average is  {weighted_avg:.2f}')
```
