```python
def crop_data(filename):
  crops=[]
  file=open(filename,"r")
  for line in file:
    line=line.strip()
    if line:
      data=line.split(",")
      crop={"name":data[0],"season":data[1],"area":float(data[2]),"yield":float(data[3]),"cost":float(data[4]),"revenue":float(data[5])}
      crops.append(crop)
  file.close()
  return crops
new_crops=crop_data("crops_profit1.txt")
print(new_crops)


def find_kharif_and_non_kharif_crop(crops):
  kharif=[]
  non_kharif=[]
  for crop in crops:
    if crop["season"]=="Kharif":
      kharif.append(crop["name"])
    else:
      non_kharif.append(crop["name"])
  return "kharif_crops",kharif,"non kharif crops",non_kharif
crops=crop_data("crops_profit1.txt")
fine_crops=find_kharif_and_non_kharif_crop(crops)
print(fine_crops)

def crops_profit(crops):
  for crop in crops:
    profit=crop["revenue"]-crop["cost"]
    print(crop["name"],":","profit",profit)

crops_profit(crops)

def summary_statistics(crops):
  area=0
  for crop in crops:
    area+=crop["area"]
  yields=0
  for crop in crops:
    yields+=crop["yield"]
  avg_yield=yields/len(crops)
  return area,avg_yield
summary=summary_statistics(crops)
print("total area and avg yield;",summary)

def calculate_yield(crop):
  return crop["yield"]/crop["area"]

def crops_profit(crops):
  total_profit=[]
  for crop in crops:
    profit=crop["revenue"]-crop["cost"]
    crop_profit=[]
    crop_profit.append(crop["name"])
    crop_profit.append(profit)
    y=tuple(crop_profit)
    total_profit.append(y)
  return total_profit

def highest_yield(crops):
  highest=crops[0]
  for crop in crops:
    if calculate_yield(crop)>calculate_yield(highest):
      highest=crop
  return highest

total_crops_profits=crops_profit(crops)
print(total_crops_profits)

highest=highest_yield(crops)
print("Highest average yielding crop:",highest["name"])


```
