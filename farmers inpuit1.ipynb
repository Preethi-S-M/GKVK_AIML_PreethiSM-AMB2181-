```python
print("Farmer input based profit analysis")
seed_cost=float(input("Enter seed cost(INR):"))
fertilizer_cost=float(input("Enter fertilizer cost(INR):"))
pesticide_cost=float(input("Enter pesticide cost(INR):"))
labour_cost=float(input("Enter labour cost(INR):"))
irrigation_cost=float(input("Enter irrigation cost(INR):"))

total_input_cost=seed_cost+fertilizer_cost+pesticide_cost+labour_cost+irrigation_cost
print("\nTotal input cost:",total_input_cost)

total_yield=float(input("Enter total crop yield(kg):"))
price_per_kg=float(input("Enter price per kg(INR):"))

total_revenue=total_yield*price_per_kg
print("\nTotal revenue:",total_revenue)

total_profit=total_revenue-total_input_cost
print("\nTotal profit:",total_profit)

profit_per_kg=total_profit/total_yield
print("\nProfit per kg:",profit_per_kg)

total_seed_kg=float(input("Enter total seed (kg):"))
seed_bag_capacity=float(input("Enter seed bag capacity(kg):"))

full_seed_bag=total_seed_kg//seed_bag_capacity
remaining_seed=total_seed_kg%seed_bag_capacity

print("\nTotal full seed bags:",full_seed_bag)
print("Remaining seed(kg):",remaining_seed)

initial_crop_unit=float(input("\nEnter initial crop units:"))
growth_rate=float(input("Enter growth rate(eg:1.10 fro 10%):"))
seasons=int(input("\nEnter number of cropping seasons:"))

expected_crop=initial_crop_unit*(growth_rate**seasons)
print("\nExpected crop after season:",expected_crop)

weather_loss=float(input("Enter loss due to weather(INR)(enter positive value):"))
weather_loss=-weather_loss

adjusted_profit=total_profit+weather_loss
print("\nAdjusted profit after weather loss:",adjusted_profit)

positive_view=+adjusted_profit
print("\nPositive view of profit:",positive_view)

negative_view=-adjusted_profit
print("\nNegative view of profit:",negative_view)

print("\n------End of farmer analysis------")
print("bye 👋")
```
