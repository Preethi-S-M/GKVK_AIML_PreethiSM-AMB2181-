```python
a=10
print("Initial value:",a)

a+=5
print("afetr+=:",a)

a-=3
print("After -=",a)

a*=2
print("After *=:",a)

a/=4
print("After /=:",a)

a//=2
print("After //=:",a)

a%=3
print("After %=:",a)

a**=2
print("After **=:",a)
```

```python
x=20
y=15

print("x==y:",x==y)
print("x!=y:",x!=y)
print("x>y:",x>y)
print("x<y:",x<y)
print("x>=y:",x>=y)
print("x<=y:",x<=y)
```

```python
rainfall=True
soil_moisture=False

print("AND:",rainfall and soil_moisture)
print("OR:",rainfall or soil_moisture)
print("NOT:",not rainfall)
```

```python
a=[1,2,3]
b=a
c=[1,2,3]

print("a is b:",a is b)
print("a is c:",a is c)
print("a is not c:",a is not c)
```

```python
crops=["maize","wheat","rice"]
print("Rice in crops:","rice" in crops)
print("barley in crops:","barley"in crops)
print("barley is not in crops:","barley" not in crops)
```

```python
a=12
b=8

print("Bitwise AND(&):",a&b)
print("Bitwise OR(|):",a|b)
print("Bitwise XOR(^):",a^b)
print("Bitwise NOT(~a):",~a)
print("Left shift(a<<1):",a<<1)
print("Right shift(a>>1):",a>>1)
```
