```python
import pandas as pd
data={"Department":["HR","Sales","IT","Sales","IT","HR","Sales"],
      "Employee":["A","B","C","D","E","F","G"],
      "Salary":[50000,60000,45000,48000,70000,72000,71000],
      "Experience":[2,4,3,5,6,7,5]}
df=pd.DataFrame(data)
print(df)

```

```python
df3=df.groupby("Experience").sum()
print(df3)
```

```python
# agregation function
dfa=df.groupby("Department")["Salary"].sum()
print(dfa)
df3=df.groupby("Experience")["Salary"].sum()
print(df3)
```

```python
df1=df.groupby(["Experience","Department"]).sum()
dfq=df.groupby("Department").sum()
dfw=df.groupby(["Department","Experience"]).sum()
print(dfw)
print(dfq)
print(df1)
```

```python
# Custom aggregation
s=df.groupby("Department").agg({"Salary":"sum","Experience":"mean"})
s["Experience"] = s["Experience"].round(2)
print(s)
```

```python
df.groupby("Department")["Salary"].agg(["sum","mean","std"])
```

```python
s=df.groupby("Department")["Salary"].agg(["sum","mean","std"])
s["sum"]=s["sum"].round(2)
s["mean"]=s["mean"].round(2)
s["std"]=s["std"].round(2)
print(s)

```

```python
# broadcasting group mean to each row
df9=df.groupby("Department")["Salary"].transform("mean")
print(df9)
df["Dept_avg_salary"]=df9
print(df)
```

```python
# broadcasting group mean to each row
df["Dept_avg_salary"]=df.groupby("Department")["Salary"].transform("mean")
print(df)
```

```python
# filtering groups
df5=df.groupby("Department").filter(lambda x:len(x)>2)
print(df5)
```

```python
# pivot table
df.pivot_table(values="Salary",index="Department",columns="Experience",aggfunc="mean")
```

```python
# pivot table
df.pivot_table(values="Experience",index="Department",aggfunc="sum")
```

```python
# agricultural data example
raw_data={"Season":["Kharif","Rabi","Rabi","Summer","Kharif","Rabi","Summer","Summer"],
          "Crops":["Rice","Wheat","Blackgram","Sugarcane","Maize","Mustard","Tomato","Chilli"],
          "Area(in ac)":[3,2,1.5,1,3.1,2,2.1,2],
          "Yield(in q)":[40,30,10,60,35,12,50,50]}
df=pd.DataFrame(raw_data)
print(df)
```

```python
df.groupby(["Season","Crops"]).sum()
```

```python
df.groupby("Season")["Yield(in q)"].sum()
```

```python
df.groupby("Crops").agg({"Area(in ac)":"sum","Yield(in q)":"mean"})
```

```python
df.groupby("Season").agg({"Area(in ac)":"sum","Yield(in q)":"mean"})
```

```python
# broadcasting
avg=df["Average yields"]=df.groupby("Season")["Yield(in q)"].transform("mean")
print(avg)
print(df)
```
