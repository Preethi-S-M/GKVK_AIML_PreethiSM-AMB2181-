```python
# value is returned
def water_needs(plant_type):
  if plant_type=="corn":
    return 30
  elif plant_type=="wheat":
    return 20
  else:
    return 10
water_needs("corn")
```

```python
# two arguments passed and no value is returned
def irrigation_needed(moisture_level,rainfall):
  if moisture_level<=40 and rainfall<=10:
    print("irrigation is needed")
  else:
    print("irrigation is not needed")
irrigation_needed(40,10)
```

```python
# calculator function
# function for addition
def add(x,y):
  return x+y
# function fro subtraction
def subtract(x,y):
  return x-y
# function for multiplication
def multiply(x,y):
  return x*y
# function for division
def divide(x,y):
  if y==0:
    print("error, cannot be divided by 0")
  else:
    return x/y
def exponent(x,y):
  return x**y
# function to take values from user
def get_input():
  try:
    x=float(input("enter the first number"))
    y=float(input("enter the second number"))
    return x,y
  except ValueError:
    print("invalid input, please enter a number")
    return get_input()
# functions to print menu operations
def print_menu():
  print("\nselect operation")
  print("1.addition")
  print("2.subtraction")
  print("3.multiplication")
  print("4.division")
  print("5.square")
  print("6.exit")
# main calculator function
def calculator():
  while True:
    print_menu()
    choice=input("Enter the choice (1/2/3/4/5/6):")
    if choice=='6':
      print("exiting the calculator")
      break
    x,y=get_input()
    if choice=='1':
      print(f"{x}+{y}={add(x,y)}")
    elif choice=='2':
      print(f"{x}-{y}={subtract(x,y)}")
    elif choice=='3':
      print(f"{x}*{y}={multiply(x,y)}")
    elif choice==4:
      print(f"{x}/{y}={divide(x,y)}")
    elif choice==5:
      print(f"{x}**{y}={exponent(x,y)}")
    else:
      print("invalid choice")
calculator()
print("done with the program")


```
