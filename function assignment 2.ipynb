```python

#daily expense management
def expenses_made():
  expenses=[]
  number_of_days=int(input("Enter the number of days:"))
  for i in range(number_of_days):
    day_name=input("Enter the name of the day:")
    seed_cost=int(input("Enter the cost of the seed:"))
    fertilizer_cost=int(input("Enter the cost of the fertilizer:"))
    labour_cost=int(input("Enter the cost of the labour:"))
    expenses.append([day_name,[seed_cost,fertilizer_cost,labour_cost]])
  return expenses
total_expenses=expenses_made()
for expenditure in total_expenses:
  print(expenditure[0],"daily expenses",expenditure[1],"\nTotal expenses",sum(expenditure[1]))
grand_total=0
for expenditure in total_expenses:
  grand_total+=sum(expenditure[1])
print("grand total",grand_total)
def crop_details():
 info=[]
 number_of_crops=int(input("Enter the number of crops:"))
 for i in range(number_of_crops):
    crop_name=input("Enter the name of crop")
    season=input("Enter the season")
    yields=int(input("Enter the yield"))
    info.append((crop_name,season,yields))
 return info
y=tuple(crop_details())
print(y)
for a,b in enumerate(y,start=1):
 print(a,b )







```
