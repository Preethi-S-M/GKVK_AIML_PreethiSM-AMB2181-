```python
import pandas as pd
import numpy as np
df=pd.DataFrame({"Name":["Alice","Bob","Charlie","David","Eva","Frank"],
                 "Department":["HR","IT","IT","Finance","HR","Finance"],
                 "Salary":[50000,70000,70000,60000,52000,60000],
                 "Experience":[2,5,4,6,3,6]})
print(df)
```

```python
# sort by one column (ascending)
sal=df.sort_values("Salary")
print(sal)
```

```python
# sort by names
sal=df.sort_values("Department")
print(sal)
```

```python
# multiple columns
sal=df.sort_values(["Experience","Salary"],ascending=[False,True])
print(sal)
```

```python
# multiple columns
sal=df.sort_values(["Salary","Experience"],ascending=[False,True])
print(sal)
```

```python
# sort by index
des=df.sort_index()
print(des)
```

```python
# Ranking example
# BAsic ranking
df["Rank_salary"]=df["Salary"].rank()
print(df)
```

```python
# Ranking example
# BAsic ranking
df["Rank_salary"]=df["Salary"].rank(ascending=False)
print(df)
```

```python
# Dense ranking
df["Dense_rank_salary"]=df["Salary"].rank(ascending=False,method="dense")
print(df)
```

```python
# Value count example
df=df["Department"].value_counts()
print(df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt # Changed import to access plt directly

df_with_nan=pd.DataFrame({"Name":["Alice","Bob","Charlie","David","Eva","Frank"],
                 "Department":["HR","IT","IT","Finance",np.nan,"Finance"],
                 "Salary":[50000,70000,np.nan,60000,52000,60000],
                 "Experience":[2,5,4,6,3,np.nan]})

df_with_nan.isna()
df_with_nan.isna().sum()
```

```python
df1=pd.DataFrame({"Name":["A","B","C","D","E","F","G"],"Age":[10,9,np.nan,11,8,9,10],"Marks":[80,90,99,np.nan,np.nan,np.nan,70]})
print(df1)
df1.dropna()
```

```python
df1.dropna(axis=1) # removes columns with nan
```

```python
df1.dropna(how="all")
# removes all rows where each value is nan
```

```python
df1.dropna(thresh=3)# threshold represents how many values should be there to keep that row
# here thresjold=3 means atleast there should be 3 non missing i.e true values should be there
```

```python
df1.dropna(subset=["Marks"])
df1.dropna(thresh=3)
```

```python
# dropping only one column
df1.dropna(subset=["Marks"])
```

filling missing values


```python
#filling constant value
df1.fillna(0)
```

```python
# filling mean
df1["Marks"].fillna(df1["Marks"].mean())
```

```python
# backward fill
print(df1)
print(df1.bfill())
```

```python
# changing 2 columns
df1["Age"]=df1["Age"].ffill()
df1["Marks"]=df1["Age"].bfill()
print(df1)
```

```python
# interpolating
df2=pd.DataFrame({"Name":["A","B","C","D","E","F"],
                  "Height":[5.8,5.2,5.5,5.9,np.nan,5.6],
                  "Weight":[55,np.nan,49,58,72,80]})
df2.interpolate()
```
