```python
# index( ) method
# returns the index of the first occurence of an element

thislist=["apple","banana","cherry","banana"]
index_value=thislist.index("banana")
print("Index value of banana:", index_value)
```

```python
# count( method)
#counts how many times an element has appesred in the list

thislist=["apple","banana","apple","cherry","apple"]
count_value=thislist.count("apple")
print("count value of apple:",count_value)
```

```python
#reverse method
#reverse the order of the element in the list (in place)

thislist=["apple","mango","kiwi","banana"]
print("original list:", thislist)
thislist.reverse()
print("reversed list:", thislist)
```

```python
#reversed()

original_list=[1,2,3,4,5]
print("original list:",original_list)
reversed_iterator=reversed(original_list)
reversed_list=list(reversed_iterator)
print("reversed list:",reversed_list)
```

```python
#list length
#to determine how many items a list has len()

this_list=["apple","banana","cherry"]
print(len(this_list))
print("list length:",len(this_list))
```

```python
#list items
# list items can be of any data type

list1=["apple","banana","cherry"]
list2=[1,2,3,4,5]
list3=[True,False,False]
print(list1)
print(list2)
print(list3)
```

```python
# using len to access elements

list1=["apple","banana","cherry"]
for item in range(len(list1)):
  print(item)


# to print the list item
list1=["apple","banana","cherry"]
for item in range(len(list1)):
  print(list1[item])
```

```python
#type
#in python lists are defined as objects with data types

x=10
y=11.3
mylist=["apple","banana","cherry"]
print(type(x))
print(type(y))
print(type(mylist))
```

```python
# sort
# to arrange elemets is ascending order

number=[32,5,2,10,1,]
sorted_number=sorted(number)
sorted_list=list(sorted_number)
print("original list:",number)
print("sorted list:",sorted_list)
```
