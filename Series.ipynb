```python
import pandas as pd
mydataset=[["BMW","Volvo","Ford"],[3,4,7]]
myvar=pd.Series(mydataset)
print(myvar)
print("\n\n")
ourvar=pd.DataFrame(mydataset)
print(ourvar)
print("\n")
myvar=pd.Series(mydataset,index=["x","y"])
print(myvar)
print("\n")
ourvar=pd.DataFrame(mydataset,index=["x","y"],columns=["high","rich","poor"])
print(ourvar)
```
