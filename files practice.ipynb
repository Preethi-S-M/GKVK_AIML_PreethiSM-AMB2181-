```python

# Agricultural data analysis project
# function definition
def loan_data(filename):
  crops=[]
  with open(filename,"r") as file:
   for line in file:
    line=line.strip()
    if line:
      data=line.split(",")
      crop={
            "name":data[0],
            "season":data[1],
            "area":float(data[2]),
            "rainfall":float(data[3]),
            "fertilizer":float(data[4]),
            "yield":float(data[5])                    }
      crops.append(crop)
  return crops

def calculate_yield_per_area(crop):
  return crop["yield"]/crop["area"]

def display_crops(crops):
  print("\n Crop records")
  for crop in crops:
   ypa=calculate_yield_per_area(crop)
   print(crop["name"],"-",crop["season"],"-","yield/area:",round(ypa,2))


def find_highest_and_lowest_yield(crops):
  highest=crops[0]
  lowest=crops[0]
  for crop in crops:
    if calculate_yield_per_area(crop) >calculate_yield_per_area(highest):
      highest=crop
    if calculate_yield_per_area(crop) <calculate_yield_per_area(lowest):
      lowest=crop
  return highest,lowest

def highest_rainfall_season(crops):
  kharif_rainfall=[]
  rabi_rainfall=[]
  annual_rainfall=[]
  for crop in crops:
    if crop["season"]=="Kharif":
      kharif_rainfall.append(crop["rainfall"])
    elif crop["season"]=="Rabi":
      rabi_rainfall.append(crop["rainfall"])
    else:
      annual_rainfall.append(crop["rainfall"])
  k_avg=sum(kharif_rainfall)/len(kharif_rainfall)
  r_avg=sum(rabi_rainfall)/len(rabi_rainfall)
  a_avg=sum(annual_rainfall)/len(annual_rainfall)

  return k_avg,r_avg,a_avg

def crop_count_per_season(crops):
  count={}
  for crop in crops:
    season=crop["season"]
    if season not in count:
      count[season]=0
    count[season]+=1
  for season,value in count.items():
    print(season,":",value)
  return count

def rainfall_efficiency_analysis(crops,max_rainfall):
  print("crops with rainfall<=",max_rainfall)
  for crop in crops:
    if crop["rainfall"]<=max_rainfall:
      print(crop["name"],"-",crop["yield"])

def find_max_rainfall(crops):
  max_rain=crops[0]["rainfall"]
  for crop in crops:
    if crop["rainfall"]>max_rain:
      max_rain=crop["rainfall"]
  return max_rain


def fertilizer_efficiency(crop):
  return crop["yield"]/crop["fertilizer"]

def most_fertilizer_efficient_crop(crops):
  best=crops[0]
  for crop in crops:
    if fertilizer_efficiency(crop)>fertilizer_efficiency(best):
      best=crop
  print("most efficient crop:",best["name"])
  return best

def rank_crops_by_yield(crops):
  ranked=sorted(crops,key=calculate_yield_per_area,reverse=True)
  rank=1
  for crop in ranked:
    print(rank,":",crop["name"],"-",round(calculate_yield_per_area(crop),2))
    rank+=1
  return ranked

def best_crop_per_season(crops):
  best={}
  for crop in crops:
    season=crop["season"]
    if season not in best:
      best[season]=crop
    else:
      if calculate_yield_per_area(crop)>calculate_yield_per_area(best[season]):
        best[season]=crop
    for season,crop in best.items():
      print(season,":",crop["name"],"-",round(calculate_yield_per_area(crop),2))
  return best

def crops_in_multiple_seasons(crops):
  crop_seasons={}
  for crop in crops:
    name=crop["name"]
    season=crop["season"]

    if name not in crop_seasons:
      crop_seasons[name]=[]
    if season not in crop_seasons[name]:
     crop_seasons[name].append(season)

  for name,seasons in crop_seasons.items():
    if len(seasons)>1:
      print(name,"grown in seasons",seasons)
  return crop_seasons

def recommended_crops(crops,season,max_rainfall,max_fertilizer):
  print("Recommended crops:")
  for crop in crops:
    if crop["season"]==season and crop["rainfall"]<=max_rainfall and crop["fertilizer"]<=max_fertilizer:
      print(crop["name"])
  return recommended_crops
def seasonal_summary(crops):
  summary={}
  for crop in crops:
    season=crop["season"]
    ypa=calculate_yield_per_area(crop)
    if season not in summary:
      summary[season]=[]
    summary[season].append(ypa)
  return summary
def write_report(filename,crops,highest,lowest,k_avg,r_avg,a_avg,best,count,max_rainfall,max_fertilizer,ranked,best_crop,crop_seasons,recomended_crops,summary):
  file=open(filename,"w")
  file.write("Agricultural data analysis\n\n")
  file.write("crop yield per unit area\n")
  for crop in crops:
    ypa=calculate_yield_per_area(crop)
    file.write(crop['name']+":"+crop['season']+":"+str(round(ypa,2)) + "\n")

  file.write("\n Highest yielding crop:\n")
  file.write(highest["name"]+"\n")

  file.write("\n Lowest yielding crop:\n")
  file.write(lowest["name"]+"\n")

  file.write("\n Average rainfall for each season:\n")
  file.write("Kharif: " + str(round(k_avg, 2)) + "\n")
  file.write("Rabi: " + str(round(r_avg, 2)) + "\n")
  file.write("Annual: " + str(round(a_avg, 2)) + "\n")

  file.write("most fertilizer efficient crop:\n")
  file.write(best["name"]+"\n")

  file.write("crop count per season:\n")
  for season,value in count.items():
    file.write(season + ": " + str(value) + "\n")

  file.write("\ncrops high yielding under low rainfall:\n")
  file.write(crop["name"]+":"+str(max_rainfall) + "\n")

  file.write("\nrainfall efficiency analysis:\n")
  for crop in crops:
    if crop["rainfall"]<=max_rainfall:
      file.write(crop["name"]+":"+str(crop["yield"]) + "\n")

  file.write("\nRanked crops by yield:\n")
  for crop in ranked:
    file.write(crop["name"]+":"+str(round(calculate_yield_per_area(crop),2)) + "\n")

  file.write("\nBest crop per season:\n")
  for season,crop in best_crop.items():
    file.write(season + ": " + crop["name"] + "\n")

  file.write("\nCrops in multiple seasons:\n")
  for name,seasons in crop_seasons.items():
    if len(seasons)>1:
      file.write(name + ": " + ", ".join(seasons) + "\n")

  file.write("\nRecommended crops:\n")
  for crop in crops:
    if crop["season"]==season and crop["rainfall"]<=max_rainfall and crop["fertilizer"]<=max_fertilizer:
      file.write(crop["name"] + "\n")

  file.write("\n Seasonal summary:\n")
  for season in summary:
        avg = sum(summary[season]) / len(summary[season])
        file.write(season + ": " + str(round(avg, 2)) + "\n")

  file.close()

  # main programme starts here
def main():
  input_file="crops new.txt"
  output_file="analysis report.txt"
  crops=loan_data(input_file)
  for crop in crops:
    print(crop)
  for crop in crops:
    for key,value in crop.items():
      print(f"{key},:,{value}")
    print("-----------")
  display_crops(crops)

  highest,lowest=find_highest_and_lowest_yield(crops)

  k_avg,r_avg,a_avg=highest_rainfall_season(crops)

  best=most_fertilizer_efficient_crop(crops)

  count=crop_count_per_season(crops)

  max_rain=find_max_rainfall(crops)

  rainfall_efficiency_analysis(crops,max_rain)

  ranked=rank_crops_by_yield(crops)

  best_crop=best_crop_per_season(crops)

  crop_seasons=crops_in_multiple_seasons(crops)

  max_fertilizer=500

  recommended_crops(crops,"Kharif",max_rain,max_fertilizer)

  summary=seasonal_summary(crops)

  write_report(output_file,crops,highest,lowest,k_avg,r_avg,a_avg,best,count,max_rain,max_fertilizer,ranked,best_crop,crop_seasons,recommended_crops,summary)

  print("\n analysis report complete ,report written to :",output_file)
main() # Call the main function to execute the program
```
