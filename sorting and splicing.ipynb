```python
import numpy as np
# Crop yields for 4 farms over 5years
yields=np.array([[45,48,52,49,53],
                [38,41,39,44,42],
                [52,55,58,56,60],
                [41,43,45,47,46]])
# weather impacts factor each year
weather_factors=np.array([1.0,0.95,1.05,0.98,1.02])
print("\n Original yields(tons/ha):")
print(yields)
print("\n Weather impact factors:",weather_factors)

print("[-2:-1]",yields[-2:-1])
print("[-2:]",yields[-2:])

# Broadcasting:adjust all farm yield
adjusted_yields=yields*weather_factors
print("\n Weather adjusted Years")
print(adjusted_yields.round(2))

# boolean indexing : find exceptional years
exceptional=adjusted_yields>55
print("\n Exceptional yields(>55):")
print(adjusted_yields[exceptional])
print(exceptional)

# conditional operation :classify yields
categories=np.where(
    adjusted_yields>50,"high",
    np.where(adjusted_yields>40,"medium","low")
)
print("\nyield categories")
print(categories)

# advanced indexing:select specific farm
selected_farms=[0,2]
print("\n Selected farms yields")
print(yields[selected_farms])

# multiconditioning filtering
good_years=adjusted_yields[(adjusted_yields>45)&(adjusted_yields<55)]
print("\n Good performing years(45-55tons):")
print(good_years)
```
