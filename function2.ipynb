```python
# functions that return value from a list
def avarage_yield(yields):
  total=sum(yields)
  return total/len(yields)
crop_yield=[1200,1350,1100,1500]
avg=avarage_yield(crop_yield)
print("crop yield:",crop_yield)
print("Average yield:" ,avg)
```

```python
# function that modifies a list
def increased_yield(yields,increment):
  for i in range(len(yields)):
    yields[i]+=increment
crop_yield=[1200,1350,1100]
increased_yield(crop_yield,100)
print(crop_yield)
```

```python
# function to modify the list and give new list
def convert_to_quintals(yields):
  result=[]
  for value in yields:
    result.append(value/100)
  return result
crop_yield=[1200,1350,1100]
quintals=convert_to_quintals(crop_yield)
print("yield in kgs:",crop_yield)
print("yield in quintals:",quintals)
```

```python
# function with list +condition
def high_yield_crops(yields,threshold):
  high=[]
  for value in yields:
    if value>=threshold:
      high.append(value)
  return high
crop_yield=[1200,900,1500,800]
print(high_yield_crops(crop_yield,1000))
```

```python
# functions using built in methods
def sort_prices(prices):
  prices.sort()
  return prices
crop_prices=[1800,1500,2200,2000]
sort_prices(crop_prices)
print(crop_prices)
```

```python
# usinf sorted
def sort_prices(prices):
  prices.sort()
  crops_sorted=sorted(prices)
  return crops_sorted
crop_prices=[1800,1500,1200,2000]
new_prices=sort_prices(crop_prices)
print("original prices:",crop_prices)
print("assorted prices:",new_prices)
```

```python
# function with list and input
def collect_yields():
  yields=[]
  n=int(input("enter the number of fields:"))
  for i in range(n):
    value=float(input("enter the yield:"))
    yields.append(value)
  return yields
data=collect_yields()
print(data)

```

```python
#function usinf len ,max and min
def yield_statistics(yields):
  print("maximum yield:",max(yields))
  print("minimum yield:",min(yields))
  print("total number of fields:",len(yields))
crop_yield=[1800,1200,3400]
yield_statistics(crop_yield)
```
