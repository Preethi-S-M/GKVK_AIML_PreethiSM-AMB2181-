```python
"""
negative indexing
means to start from the end
refers to last item -1
"""
thislist=["apple","banana","cherry"]
print(thislist[-2])

```

```python
#range of indexes
thislist=["apple","banana","cherry","orange","kiwi","melon","mango"]
print(thislist[2:5])
```

```python
# if range is not mentioned
print(thislist[:4])

#by leaving out end values
print(thislist[2:])
```

```python
# negative indices
thislist=["apple","banana","cherry","orange","kiwi","melon","mango"]
print(thislist[1:-3])
```

```python
# change item value to change value of specific element
thislist=["apple","banana","cherry"]
thislist[1]="blackcurrant"
print(thislist)
```

```python
#to change the range of elements
thislist=["apple","banana","cherry"]
thislist[1:2]=["blackcurrant","watermelon"]
print(thislist)
```

```python
#append items
thislist  = ["apple","banana","cherry"]
thislist.append("orange")
print(thislist)
```

```python
#extend list
thislist=["apple","banana","cherry"]
tropical=["mango","pineapple","papaya"]
thislist.extend(tropical)
print(thislist)
```
