```python
# Advanced scatter plot
import matplotlib.pyplot as plt
import numpy as np

x=np.random.rand(50)
y=np.random.rand(50)
sizes=1000*np.random.rand(50)
colors=np.random.rand(50)

plt.scatter(x,y,s=sizes,c=colors,cmap='plasma',alpha=0.6)
plt.colorbar()
plt.show()
```

```python
#Bar graphs
import matplotlib.pyplot as plt
import numpy as np

x=np.array(["a","b","c","d"])
y=np.array([3,8,1,10])

plt.bar(x,y)
plt.grid(True)
plt.show()
```
