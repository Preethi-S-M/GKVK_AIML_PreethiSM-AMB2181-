```python
# for loop (range)
for i in range(5):
  print(i)

```

```python
for i in range(1,11,2):
  print(i)
```

```python
# iterating over a list
crops=["Rice","Wheat","Maize"]
for crop in crops:
  print(crop)
```

```python
#iterating over a string
crop="Rice"
for letter in crop:
  print(letter)
```

```python
#example
days=int(input("Enter number of days: "))
total_yield=0
for day in range(1,days+1):
  yield_today=float(input(f"Enter yield for day {day}:"))
  total_yield+=yield_today
print("Total yield:",total_yield)
```

```python
# nested for loop
for i in range(1,4):
  for j in range(1,4):
    print(i,j)
```

```python
#else with for loop
for i in range(5):
  print(i)
else:
 print("loop completed successfully")
```

```python
# for loop with break
for i in range(1,6):
  if i==4:
    break
  print(i)
```

```python
# altered
for i in range(1,6):
  print(i)
  if i==4:
    break
```

```python
#continue
for i in range(1,6):
  if i==3:
    continue
  print(i)
```

```python
#altered continue
for i in range(1,6):
  print(i)
  if i==3:
    continue
```
